```python
import os

data_path = "data/"
if not os.path.exists(data_path):
    os.makedirs(data_path)
data_path = "data/"
```

```python
resolution = 30
octaves = 4
dataset_size = 1024
drop_percs = [x/100 for x in range(0, 100, 5)]
epochs = 100
```

```python
import numpy as np

zero_filling_cnn_wces = np.load(data_path+'zero_filling_cnn_{resolution}x{resolution}_{octaves}_{dataset_size}_{epochs}_wces.npy'.format(resolution=resolution, octaves=octaves, dataset_size=dataset_size, epochs=epochs))
cnn_filling_gd_wces = np.load(data_path+'cnn_filling_gd_{resolution}x{resolution}_{octaves}_{dataset_size}_{epochs}_wces.npy'.format(resolution=resolution, octaves=octaves, dataset_size=dataset_size, epochs=epochs))
knn_filling_gd_wces = np.load(data_path+'knn_filling_gd_{resolution}x{resolution}_{octaves}_{dataset_size}_wces.npy'.format(resolution=resolution, octaves=octaves, dataset_size=dataset_size))
knn_filling_cnn_wces = np.load(data_path+'knn_filling_cnn_{resolution}x{resolution}_{octaves}_{dataset_size}_{epochs}_wces.npy'.format(resolution=resolution, octaves=octaves, dataset_size=dataset_size, epochs=epochs))
# cnn_filling_cnn_wces = np.load(data_path+'cnn_filling_cnn_{resolution}x{resolution}_{octaves}_{dataset_size}_{epochs}_wces.npy'.format(resolution=resolution, octaves=octaves, dataset_size=dataset_size, epochs=epochs))
```

```python
import matplotlib.pyplot as plt

plt.plot(drop_percs, knn_filling_gd_wces, marker='o', label="KNN filling GD")
plt.plot(drop_percs, cnn_filling_gd_wces, marker='o', label="CNN filling GD")
plt.plot(drop_percs, zero_filling_cnn_wces, marker='o', label="Zero filling CNN")
plt.plot(drop_percs, knn_filling_cnn_wces, marker='o', label="KNN filling CNN")
# plt.plot(drop_percs, cnn_filling_cnn_wces, marker='o', label="CNN filling CNN")
plt.legend(fontsize=12)
plt.xlabel("drop percentage")
plt.ylabel("weighted cross-entropy")
plt.show()
```
